```python
# âœ… Install required packages
#!pip install langchain sentence-transformers chromadb google-generativeai bs4

# âœ… Imports
import re
import requests
from bs4 import BeautifulSoup
from typing import List, Dict

# âœ… Scraper Class
class WebContentScraper:
    def fetch_text_from_url(self, url: str) -> str:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Remove script/style
        for s in soup(['script', 'style']):
            s.decompose()

        return soup.get_text()

    def preprocess_text(self, text: str) -> str:
        text = re.sub(r'\s+', ' ', text)
        return text.strip()

    def chunk_text(self, text: str, chunk_size: int = 500) -> List[str]:
        words = text.split()
        return [' '.join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]

# âœ… Chatbot Class
class GeminiRAGChatbot:
    def __init__(self, api_key: str):
        from google.generativeai import configure, GenerativeModel
        from langchain_community.embeddings import HuggingFaceEmbeddings


        configure(api_key=api_key)
        self.model = GenerativeModel("models/gemini-1.5-flash")
        self.embedding_model = HuggingFaceEmbeddings()
        self.scraper = WebContentScraper()
        self.vector_store = None
        self.conversation_history = []

    def initialize_knowledge_base(self, url: str) -> bool:
        try:
            print("ğŸŒ Scraping website content...")
            content = self.scraper.fetch_text_from_url(url)

            print("ğŸ”„ Preprocessing and chunking text...")
            clean_content = self.scraper.preprocess_text(content)
            chunks = self.scraper.chunk_text(clean_content)

            print("ğŸ§  Embedding and storing chunks in vector store...")
            from langchain.vectorstores import Chroma
            from langchain.schema import Document

            documents = [Document(page_content=chunk) for chunk in chunks]
            self.vector_store = Chroma.from_documents(
            documents,
            embedding=self.embedding_model,
            persist_directory=None  # âœ… Ensures it's in-memory and resets every time
)

            return True
        except Exception as e:
            print(f"âŒ Error during knowledge base initialization: {str(e)}")
            return False

    def generate_response(self, query: str, context_chunks: List[str]) -> str:
        context = "\n\n".join([f"Context {i+1}:\n{chunk}" for i, chunk in enumerate(context_chunks)])
        prompt = f"""You are an AI assistant that helps answer questions about LLM-powered autonomous agents based on the provided context.

Use the following context to answer the user's question. If the context doesn't contain enough information to answer the question, say so politely and provide what information you can.

CONTEXT:
{context}

USER QUESTION: {query}

Please provide a comprehensive answer based on the context above. If you reference specific concepts or techniques, explain them clearly."""

        try:
            response = self.model.generate_content(prompt)
            return response.text
        except Exception as e:
            return f"Sorry, I encountered an error generating the response: {str(e)}"

    def chat(self, query: str) -> str:
        if not query.strip():
            return "Please ask a question about LLM-powered autonomous agents."

        search_results = self.vector_store.similarity_search(query, k=3)

        if not search_results:
            return "I couldn't find relevant information in my knowledge base."

        context_chunks = [doc.page_content for doc in search_results]
        response = self.generate_response(query, context_chunks)

        self.conversation_history.append({
            'query': query,
            'response': response,
            'context_used': len(context_chunks)
        })

        return response

    def get_conversation_history(self) -> List[Dict]:
        return self.conversation_history

    def clear_history(self):
        self.conversation_history = []

# âœ… Interactive CLI Interface
def create_interactive_chat_interface(chatbot: GeminiRAGChatbot):
    print("ğŸ¤– RAG Chatbot is ready!")
    print("ğŸ’¡ Ask questions about LLM-powered autonomous agents")
    print("ğŸ”„ Type 'quit' to exit, 'clear' to clear history")
    print("-" * 50)

    while True:
        user_input = input("\nğŸ§‘ You: ").strip()

        if user_input.lower() in ['quit', 'exit', 'q']:
            print("ğŸ‘‹ Goodbye!")
            break

        if user_input.lower() in ['clear', 'reset']:
            chatbot.clear_history()
            print("ğŸ—‘ï¸ Conversation history cleared!")
            continue

        if user_input.lower() in ['help', 'h']:
            print("""
Available commands:
- Ask any question about content you have given
- 'quit' or 'exit' - Exit the chat
- 'clear' or 'reset' - Clear conversation history
- 'help' or 'h' - Show this help message
            """)
            continue

        if not user_input:
            continue

        print("ğŸ¤– Assistant: ", end="")
        response = chatbot.chat(user_input)
        print(response)

# âœ… Main Function
def main():
    print("ğŸ”— Please enter the website URL to build the knowledge base from:")
    website_url = input("Enter website URL (e.g., https://example.com): ").strip()

    if not website_url:
        print("âŒ Website URL is required!")
        return

    # Hardcoded Gemini API key (ONLY FOR LOCAL TESTING)
    api_key = "AIzaSyAHDSuswubSqFvx0r5fHizWgFl2AHMeRaE"

    chatbot = GeminiRAGChatbot(api_key)

    success = chatbot.initialize_knowledge_base(website_url)

    if not success:
        print("âŒ Failed to initialize knowledge base")
        return

    create_interactive_chat_interface(chatbot)


# âœ… Run
if __name__ == "__main__":
    print("ğŸš€ RAG Chatbot Setup")
    print("=" * 30)

    try:
        import google.colab
        print("ğŸ” Google Colab detected")
    except ImportError:
        pass

    main()
```
